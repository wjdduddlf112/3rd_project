```python
import pandas as pd
```

```python
restaurant_df = pd.read_csv("db_csv_preprocessed/restaurants_with_coords.csv")
menu_df = pd.read_csv("db_csv_preprocessed/menu_df.csv")
review_df = pd.read_csv("db_csv_preprocessed/review_df.csv")
```

## restaurant_df 구조 분석

```python
restaurant_df.head()
```

```python
tags = restaurant_df["tags"].unique()
tags = [t for t in tags if not pd.isna(t)]
buff = []
for t in tags:
    sublist = t.split(",")
    for s in sublist:
        buff.append(s)

tags = list(set(buff))
print(len(tags))
print(tags)
```

총 37개의 unique value가 존재하는 tag들

```python
characters = restaurant_df["character"].unique()
characters = [c for c in characters if not pd.isna(c)]
buff = []
for c in characters:
    sublist = c.split(",")
    for s in sublist:
        buff.append(s)

characters = list(set(buff))
print(len(characters))
print(characters)
```

characters는 76개의 unique value가 존재

```python
shared = set(tags) & set(characters)
print(shared)
```

tag와 character의 역할은 비슷하지만 겹치는 값은 존재하지 않는 것으로 보임

## review_df 구조 분석

```python
review_df.head()
```

```python
keywords = review_df["keywords"].unique()
keywords = [k for k in keywords if not pd.isna(k)]
buff = []
for k in keywords:
    sublist = k.split(",")
    for s in sublist:
        buff.append(s)

keywords = list(set(buff))
print(len(keywords))
print(keywords)
```

```python
print(set(keywords) & set(tags))
print(set(keywords) & set(characters))
```

리뷰의 keyword와 tag, characteristic간에는 겹치는 값들이 확인됨

## category 테이블의 작성

```python
categories = restaurant_df["category"].unique()
categories = [k for k in categories if not pd.isna(k)]
buff = []
for k in categories:
    sublist = k.split(",")
    for s in sublist:
        buff.append(s)

categories = list(set(buff))
print(len(categories))
print(categories)
```

```python
category_df = pd.DataFrame(categories, columns=["category_name"])
category_df.head()
```

```python
category_df.to_csv("db_csv_cat_tag/category.csv", index=False)
```

## tag 테이블의 작성

tag, character, 리뷰의 keyword가 비슷한 역할을 가졌으니 병합해서 tag 키워드로 관리

추가로 restaurant <> keyword 간의 N:M 관계 테이블을 정의

```python
summed_tags = set(keywords) | set(tags) | set(characters)
print(len(summed_tags))
```

```python
tag_df = pd.DataFrame(summed_tags, columns=["tag_name"])
tag_df.head()
```

```python
tag_df.to_csv("db_csv_cat_tag/tag.csv", index=False)
```
