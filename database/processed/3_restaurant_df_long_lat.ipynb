```python
import pandas as pd

# 현재 위치가 이미 'processed'이므로 그 하위 폴더부터 시작
input_path = 'db_csv_preprocessed/restaurant_df.csv'
output_path = 'db_csv_preprocessed/restaurants_with_coords.csv'

try:
    df = pd.read_csv(input_path)
    df.to_csv(output_path, index=False, encoding='utf-8-sig')
    print("✅ 처리가 완료되었습니다.")
except Exception as e:
    print(f"❌ 에러 발생: {e}")
```
